```python
import numpy as np
import matplotlib.pyplot as plt

# Günlük siparişlerin normal dağılım parametreleri
mean_orders = 100  # Ortalama sipariş sayısı
std_dev_orders = 15  # Standart sapma

# Monte Carlo simülasyonu (örneğin 10.000 gün için)
num_simulations = 1000
simulated_orders = np.random.normal(mean_orders, std_dev_orders, num_simulations)

# Simülasyon sonuçlarını analiz et ve grafiği çiz
plt.hist(simulated_orders, bins=30, edgecolor='black')
plt.title("Monte Carlo Simülasyonu ile Günlük Sipariş Sayısı Tahmini")
plt.xlabel("Günlük Sipariş Sayısı")
plt.ylabel("Frekans")
plt.show()

# Simülasyon sonuçları için ortalama ve güven aralığı hesapla
average_simulated_orders = np.mean(simulated_orders)
confidence_interval = np.percentile(simulated_orders, [2.0, 95.5])

print(f"Simülasyon sonucu ortalama günlük sipariş sayısı: {average_simulated_orders:.2f}")
print(f"%95 güven aralığı: {confidence_interval}")
```
